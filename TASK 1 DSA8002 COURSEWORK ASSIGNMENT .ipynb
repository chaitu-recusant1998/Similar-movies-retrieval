```python
import pandas as pd
import numpy as np

# Importing the movie dataset 

data_movies= pd.read_csv('movies.csv')
display(data_movies)

# Cleaning the movies.csv file

# 1-Check for Duplicates
data_movies.duplicated()
data_movies.duplicated().sum()

# 2-Droping the Duplicats values
data_movies.drop_duplicates()

# 3-Filter rows containing blanks 
# We are using isnull() method to filter out rows containing null values.
data_movies.isnull()
data_movies.isnull().sum()

# We can also use an alternative function isna() to filter out rows containing null values.
data_movies.isna()
data_movies.isna().sum()

data_movies

# The filtered data frame
print(data_movies)


```

```python
# Importing the ratings dataset 

data_rating = pd.read_csv('ratings.csv')
data_rating
# Cleaning the ratings.csv file

# 1-Check for Duplicates
data_rating.duplicated()
data_rating.duplicated().sum()

# 2-Droping the Duplicats values
data_rating.drop_duplicates()

# 3-Filter rows containing blanks 
# We are using isnull() method to filter out rows cointaning null values.
data_rating.isnull()
data_rating.isnull().sum()

# We can also use an alternative function isna() to filter out rows cointaning null values.
data_rating.isna()
data_rating.isna().sum()

# The filtered data frame
print(data_rating)
data_rating = data_rating.groupby("movieId")["rating"].apply(list)
data_rating
```

```python
# Importing the tags dataset 

data_tag = pd.read_csv('tags.csv')
data_tag

# Cleaning the tags.csv file

# 1-Check for Duplicates
data_tag.duplicated()
data_tag.duplicated().sum()

# 2-Droping the Duplicats values
data_tag.drop_duplicates()

# 3-filter rows containing blanks 
# We are using isnull() method to filter out rows cointaning null values.
data_tag.isnull()
data_tag.isnull().sum()

#we are using another function isna() to filter out rows cointaning null values.
data_tag.isna()
data_tag.isna().sum()

# The filtered data frame
print(data_tag)

data_tag = data_tag.groupby("movieId")["tag"].apply(list)
```

```python
#Merging all the the three datasets according to our requirement 

#Merging movie and rating dataset
data_merge1= pd.merge(data_movies,data_rating, on = 'movieId')
data_merge1
#Merging movie+rating and tag dataset
data_merge= pd.merge(data_merge1,data_tag, on = 'movieId')
data_merge

#Fixing the genres coloumn withe the help of str.split()
data_merge['genres']= data_merge['genres'].str.split('|')

```

```python
#Function to search for Similar Movie Retrieval and enhancing Similar Movie Retrieval ! 

def funct(c1,c2):
    c1 = list(c1)[0]
    flag = 0
    for i in c1:
        for j in c2:
            if(i == j):
                flag = 1

    return flag
```

```python
y= 1
while y!= 0:
    
    X=int(input("Enter the movie id: "))
    data_find= data_merge[data_merge.movieId == X]
    print("\n")
    check = X in data_merge['movieId'].unique()
    if check == 1:
        lis = []
        lis1=[]
        l=[]
        #first genres search
        for i in range(0,len(data_merge)):
            
            c2 = data_merge.iloc[i,:].genres  
            c1=data_find.genres
    
            if(funct(c1,c2)==1):
               lis.append(i)
                
        data_col=data_merge.iloc[lis,:]
        
       #rating search 
        for i in range(0,len(data_col)):
            c2 = data_col.iloc[i,:].rating
            c1=data_find.rating
    
            if(funct(c1,c2)==1):
                lis1.append(i)

        data_collect = data_col.iloc[lis1,:]           #TASK 1 (b)
    
    
              #tag based
        for i in range(0,len(data_merge)):
            c2 = data_merge.iloc[i,:].tag  
            c1=data_find.tag
    
            if(funct(c1,c2)==1):                      #TASK 1 (c)
                l.append(i)

        data_tagcollect = data_merge.iloc[l,:] 
        
        
        data_find = data_merge[data_merge.movieId == X]
        print("The movie deatials equivalent to movieId is",X,"is :")
        display(data_find)
        print("The List of Movies and their details which are similar to movie ",data_find.title.to_string(index=False),"is :")
        display(data_collect)
        print("The List of Movies and their details which are similar to movie ",data_find.title.to_string(index=False),
              "based on its tags is :" )
        display(data_tagcollect)
        
    else:
        print("The movie id ",X," is not present in the filtered dataFrame ")
        
        
       
    y= int(input("Enter 0 to exit the search  or 1 to continue : "))      

 
        
    
    



```
